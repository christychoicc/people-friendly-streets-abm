```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.font_manager as fm

files  = ['baseline.png', 'pfs.png', 'indvidualexemption.png']
labels = ['Baseline',
          'People-Friendly Streets',
          'Individual Exemption under\nPeople-Friendly Streets']

# Use Aptos if available, otherwise fall back to Calibri then sans-serif
available = [f.name for f in fm.fontManager.ttflist]
font = next((f for f in ['Aptos', 'Calibri', 'Arial'] if f in available), 'sans-serif')

fig, axes = plt.subplots(1, 3, figsize=(15, 6))
fig.patch.set_facecolor('white')

for ax, fname, label in zip(axes, files, labels):
    img = mpimg.imread(fname)
    ax.imshow(img)
    ax.axis('off')
    ax.text(0.5, -0.04, label, transform=ax.transAxes,
            fontsize=13, fontweight='normal', fontfamily=font,
            ha='center', va='top')

plt.tight_layout(pad=1.5)
plt.savefig('combined.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()
print(f'Saved combined.png  (font used: {font})')
```
